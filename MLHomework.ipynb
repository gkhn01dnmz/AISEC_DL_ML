```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import pandas as pd
import math
```

```python
df=pd.read_csv('/content/drive/MyDrive/YapayZeka/hiring.csv')
```

Some regulazition to easy coding


```python
df['test_score']=df['test_score(out of 10)']
df['interview_score']=df['interview_score(out of 10)']
df['salary']=df['salary($)']
df
```

we are convert to string number to integer. We are dropping unnecassary columns.

```python

sayi_sozlugu = {'one': 1, 'two': 2, 'three': 3, 'four': 4, 'five': 5, 'six': 6, 'seven': 7, 'eight': 8, 'nine': 9, 'ten': 10,'eleven':11}
df['experience'] = df['experience'].map(sayi_sozlugu)
df_new=df.drop(['test_score(out of 10)','interview_score(out of 10)','salary($)'],axis=1)
df_new
```

Finding median number and put them inside of nan values.

```python
exp_median=math.floor(df_new.experience.median())
```

```python
df_new.experience=df_new.experience.fillna(exp_median)
```

```python
df_new
```

```python
test_median=math.floor(df_new.test_score.median())
df_new.test_score=df_new.test_score.fillna(test_median)
df_new
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(df_new[['experience','test_score','interview_score']],df_new.salary)
```

```python
X_test = np.array([[10,8,7]])
reg.predict(X_test)
```
